문제 4. 한 DataFrame의 null 값을 다른 DataFrame의 null이 아닌 값으로 채워서 두 DataFrame 개체를 결합하여 하나로 나타내는 프로그램을 작성하세요.

```python
df1 = pd.DataFrame({'A': [None, 0, None,4], 'B': [3, 4, 5,6]})
df2 = pd.DataFrame({'A': [1, 1, 1, 3], 'B': [1,2, None, 3]})

print("Original DataFrames:")
print(df1)
print("---------------")
print(df2)
print("---------------")
```

```python
# 문제4 답.
df1 = pd.DataFrame({'A': [None, 0, None,4], 'B': [3, 4, 5,6]})
df2 = pd.DataFrame({'A': [1, 1, 1, 3], 'B': [1,2, None, 3]})

df1['A'] = np.where(pd.notnull(df1['A']) == True, df1['A'], df2['A'])

df2['B'] = np.where(pd.notnull(df2['B']) == True, df2['B'], df1['B'])

print("문제 4번 답 DataFrames:")
print(df1)
print("---------------")
print(df2)
print("---------------")
```
